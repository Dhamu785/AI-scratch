## **Flipping the images**

```python
flip_lr = np.fliplr(img_data_1)
flip_ud = np.flipud(img_data_1)

images = [img_data_1, flip_lr, flip_ud]
colors = ['gray', 'hsv','blues']
for i in range(len(images)):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i], cmap=colors[i])
    plt.axis('off')

plt.show()
```

## **Image processing using sk-image**

```python
import matplotlib.pyplot as plt
from skimage import io
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
path = "/content/drive/MyDrive/Colab Notebooks YouTube tutorials/8 Tensorflow and keras/04_DigitalSreeni/02_Image_processing/Images/rocket.png"

img_data_1 = io.imread(path, as_gray=True)
img_data_1.shape
```

```python
rescaled_1 = rescale(img_data_1, 1.0/4.0, anti_aliasing=True)
rescaled_2 = rescale(img_data_1, 1.0/4.0, anti_aliasing=False)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(rescaled_1, cmap='gray')
plt.title("Aliasing True")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(rescaled_2, cmap='gray')
plt.title("Aliasing False")
plt.axis('off')

plt.show()
```

```python
resize_1 = resize(img_data_1, (50,50), anti_aliasing=True)
resize_2 = resize(img_data_1, (50,50), anti_aliasing=False)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(resize_1, cmap='gray')
plt.title("Aliasing True")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(resize_2, cmap='gray')
plt.title("Aliasing False")
plt.axis('off')

plt.show()
```

```python
downscale_1 = downscale_local_mean(img_data_1, (4,3))
downscale_2 = downscale_local_mean(img_data_1, (5,5))

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(downscale_1, cmap='gray')
plt.title("4x3")
# plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(downscale_2, cmap='gray')
plt.title("5x5")
# plt.axis('off')

plt.show()
```
